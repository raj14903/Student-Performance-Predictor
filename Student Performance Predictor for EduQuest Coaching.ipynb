Import necessary libraries

```python
import pandas as pd
```

Read the dataset

```python
df = pd.read_csv('/content/Student Performance Predictor for EduQuest Coaching.csv')
```

Data preprocessing

```python
df.head(6)
```

```python
df.shape
```

```python
df.describe()
```

```python
# find the null data
df.isnull().sum()
```

```python
# drop the null data
df = df.dropna()
```

```python
df.info()
```

```python
# convert categotical data into numerical data
```

```python
# gender
df['gender'].value_counts()
```

```python
df['gender'] = df['gender'].replace({'Female':1, 'Male':0})
```

```python
# parental_education
df['parental_education'].value_counts()
```

```python
df['parental_education'] = df['parental_education'].replace({'Diploma':0,'High School':1,'Graduate':2,'Postgraduate':3})
```

```python
# internet_access
df['internet_access'].value_counts()
```

```python
df['internet_access'] = df['internet_access'].replace({'Yes':1, 'No':0})
```

```python
# extra_curricular_involvement
df['extra_curricular_involvement'].value_counts()
```

```python
df['extra_curricular_involvement'] = df['extra_curricular_involvement'].replace({'Low':0,'Moderate':1,'High':2})
```

```python
# tutor_support
df['tutor_support'].value_counts()
```

```python
df['tutor_support'] = df['tutor_support'].replace({'Yes':1, 'No':0})
```

Specify X and y

```python
X = df[['gender','age','parental_education','family_income','internet_access','previous_exam_score','attendance_rate',
        'homework_completion_rate','class_participation_score','number_of_absences', 'extra_curricular_involvement',
        'learning_hours_per_week','tutor_support']]
y = df['final_exam_score']
```

Dataset splitting

```python
# use the scikit-learn library
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y ,train_size=0.75, random_state=2529)
```

Model selection

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

Train the model

```python
model.fit(X_train,y_train)
```

Prediction

```python
y_pred = model.predict(X_test)
```

Evaluation

```python
# using mean_absolute_percentage_error
from sklearn.metrics import mean_absolute_percentage_error
mean_absolute_percentage_error(y_test,y_pred)
```

The MAPE value of 5.75% indicates the model is predecting the test data accurately with only a small error
